## MULTIVARIATE ANAYLYSIS

```python
from sklearn.decomposition import PCA
# Select the desired columns
data1 = df[['Age','Flight_Distance', 'Inflight_wifi_service', 'Departure_Arrival_time_convenient', 'Ease_of_Online_booking', 'Gate_location', 'Inflight_entertainment', 'On_board_service', 'Leg_room_service', 'Baggage_handling', 'Checkin_service', 'Inflight_service', 'Cleanliness', 'Departure_Delay_in_Minutes', 'Arrival_Delay_in_Minutes', 'satisfaction']]

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data1)

# Create a new DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Display the first few rows of the new DataFrame
print(principal_df.head())
```

PCA is a technique used to reduce the dimensionality of a dataset while retaining as much of the original variation as possible. The principal components are linear combinations of the original variables that capture the maximum amount of variance in the data.

The values in the PC1 and PC2 columns represent the scores of each observation on the first and second principal components, respectively. These scores can be used to identify patterns in the data and to perform further analysis. 

## Distributions

```python
# Plot a histogram of the Age column
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```

```python
# Test if the Age column is normal
p_value = stats.normaltest(df["Age"]).pvalue
if p_value < 0.05:
    print("The Age column does not follow a normal distribution (p-value={:.4f}).".format(p_value))
else:
    print("The Age column follows a normal distribution (p-value={:.4f}).".format(p_value))
```

```python
# Plot a probability plot of the Age column
stats.probplot(df["Age"], plot=plt)
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Observed Quantiles")
plt.title("Probability Plot of Age")
plt.show()
```

```python
# Transform the Age column to a normal distribution using a log transformation
df["Age_log"] = np.log(df["Age"])
plt.hist(df["Age_log"], bins=20)
plt.xlabel("Log(Age)")
plt.ylabel("Frequency")
plt.title("Histogram of Log(Age)")
plt.show()
```

the code uses the normaltest function from the scipy.stats module to test whether the “Age” column follows a normal distribution. The null hypothesis of this test is that the data follow a normal distribution. The p-value of the test is less than 0.05, which means that we can reject the null hypothesis and conclude that the “Age” column does not follow a normal distribution.

To visually assess the fit of the “Age” column to a normal distribution, the code plots a probability plot of the “Age” column using the probplot function from the scipy.stats module. The probability plot shows that the data points deviate from the straight line in the tails, which confirms that the “Age” column is not normally distributed.

Finally, the code transforms the “Age” column to a normal distribution using a log transformation. The log transformation is a common method for transforming right-skewed data to a more normal distribution. The code plots a histogram of the transformed “Age” column to show its frequency distribution. The histogram shows that the transformed “Age” column is approximately normally distributed.

## Hypothesis testing

### Test: 
####  The mean age of customers who are satisfied with the airline is different from the mean age of customers who are dissatisfied, you can use a two-sample t-test.

## performs a two-sample t-test on the “Age” column of your dataset to test whether the mean age of satisfied customers is different from the mean age of dissatisfied customers

```python
from scipy.stats import ttest_ind

# Split the dataset into two groups based on satisfaction
satisfied = df[df["satisfaction"] == "satisfied"]["Age"]
dissatisfied = df[df["satisfaction"] == "neutral or dissatisfied"]["Age"]

# Perform a two-sample t-test
t_stat, p_value = ttest_ind(satisfied, dissatisfied)

# Print the results
if p_value < 0.05:
    print("There is a significant difference in the mean age of satisfied and dissatisfied customers (p-value={:.4f}).".format(p_value))
else:
    print("There is no significant difference in the mean age of satisfied and dissatisfied customers (p-value={:.4f}).".format(p_value))
```

we can conclude that there is a significant difference in the mean age of satisfied and dissatisfied customers (p-value=0.0000). 

This means that the difference in age between satisfied and dissatisfied customers is unlikely to be due to chance alone.

However, it is important to note that statistical significance does not necessarily imply practical significance.

In other words, even though the difference in age is statistically significant, it may not be large enough to be practically meaningful.
